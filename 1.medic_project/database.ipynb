```python
import os
import glob
import sqlite3

def final_phase1_run():
    # 1. Path define karo jahan tumhari 5 photos hain
    raw_dir = r'C:\Users\PRITAM\1.medic_project\dataset\raw_images'
    
    # 2. Check karo folder exist karta hai ya nahi
    if not os.path.exists(raw_dir):
        print(f"🚨 Folder nahi mila: {raw_dir}")
        return

    # 3. Saari images ki list nikalna
    image_extensions = ['*.jpg', '*.jpeg', '*.png']
    files_to_process = []
    for ext in image_extensions:
        files_to_process.extend(glob.glob(os.path.join(raw_dir, ext)))

    if not files_to_process:
        print("🚨 Folder khali hai! Kam se kam ek photo dalo.")
        return

    print(f"🔍 Found {len(files_to_process)} images. Processing start ho raha hai...")

    # 4. Loop chalakar har image ko process karna
    for img_path in files_to_process:
        # File name se medicine name nikalna (e.g. 'dolo.jpg' -> 'dolo')
        name_from_file = os.path.basename(img_path).rsplit('.', 1)[0]
        process_new_medicine(img_path, name_from_file)

# Ise run karo
final_phase1_run()
```
